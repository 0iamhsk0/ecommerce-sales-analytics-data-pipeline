## View 1. Daily Sales Summary

```python
%sql
CREATE OR REPLACE VIEW ecommerce.gold.vw_daily_sales_performance AS
SELECT
  transaction_date,
  COUNT(DISTINCT transaction_id) AS total_transactions,
  SUM(quantity)                  AS total_items_sold,
  SUM(net_amount_inr)             AS total_revenue_inr
FROM ecommerce.gold.fact_transactions_denorm
GROUP BY transaction_date;
```

## View 2: Customer Value Summary

```python
%sql
CREATE OR REPLACE VIEW ecommerce.gold.vw_customer_value_summary AS
SELECT
  customer_id,
  COUNT(DISTINCT transaction_id) AS total_transactions,
  SUM(net_amount_inr)            AS lifetime_value_inr,
  ROUND(
    SUM(net_amount_inr) / COUNT(DISTINCT transaction_id),
    2
  ) AS avg_transaction_value
FROM ecommerce.gold.fact_transactions_denorm
GROUP BY customer_id;
```

## View 3: Channel Performance

```python
%sql
CREATE OR REPLACE VIEW ecommerce.gold.vw_channel_performance AS
SELECT
  channel,
  COUNT(DISTINCT transaction_id) AS total_transactions,
  SUM(net_amount_inr)            AS total_revenue_inr
FROM ecommerce.gold.fact_transactions_denorm
GROUP BY channel;
```

## Vie w4: Discount Impact

```python
%sql
CREATE OR REPLACE VIEW ecommerce.gold.vw_discount_impact AS
SELECT
  coupon_flag,
  COUNT(DISTINCT transaction_id) AS transactions,
  SUM(discount_amount)           AS total_discount_given,
  SUM(net_amount_inr)            AS net_revenue_inr
FROM ecommerce.gold.fact_transactions_denorm
GROUP BY coupon_flag;
```
